```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 21
x_DE_32 = np.linspace(0, 2000, 10)
y_DE_32 = np.array([100, 60, 40, 15, 10, 9, 8, 7, 6, 1])
y_std_DE_32 = np.array([21, 18, 15, 7, 5, 4, 2, 1, 1, 0.5])

x_CEM_32 = np.linspace(0, 2000, 10)
y_CEM_32 = np.array([125, 80, 65, 40, 27, 20, 15, 10, 5, 2])
y_std_CEM_32 = np.array([30, 28, 25, 17, 15, 10, 8, 7, 6, 2])

fig, ax = plt.subplots()
ax.plot(x_DE_32, y_DE_32, '-', label="DE-32", color='tab:red')
ax.fill_between(x_DE_32, y_DE_32 - y_std_DE_32, y_DE_32 + y_std_DE_32, alpha=0.2, color='tab:red')
ax.plot(x_DE_32, y_DE_32, 'o', color='tab:red')

ax.plot(x_CEM_32, y_CEM_32, '-', label="CEM-32", color='tab:blue')
ax.fill_between(x_CEM_32, y_CEM_32 - y_std_CEM_32, y_CEM_32 + y_std_CEM_32, alpha=0.2, color='tab:blue')
ax.plot(x_CEM_32, y_CEM_32, 'o', color='tab:blue')

plt.xlabel("Number of fitness evaluations")
plt.ylabel("Objective function value")
plt.legend()
```
